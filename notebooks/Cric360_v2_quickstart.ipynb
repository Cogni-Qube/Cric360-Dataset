## 2. Load the `grounds` subset (real broadcast frames)

```python
from datasets import load_dataset

grounds = load_dataset("sarimshahzad/Cric-360", "grounds", split="train")
print(grounds)
print("\nColumns:", grounds.column_names)
```

## 3. Peek at the metadata

```python
import pandas as pd

df = grounds.remove_columns(["image"]).to_pandas()
df.head()
```

## 4. Show some sample frames

```python
import matplotlib.pyplot as plt

n = 6
idxs = list(range(0, len(grounds), max(1, len(grounds) // n)))[:n]
fig, axes = plt.subplots(2, 3, figsize=(16, 7))
for ax, i in zip(axes.ravel(), idxs):
    row = grounds[i]
    ax.imshow(row["image"])
    ax.set_title(f"{row['source_tag']} · {row['quality']} · {row['width']}x{row['height']}", fontsize=10)
    ax.axis("off")
plt.tight_layout()
plt.show()
```

## 5. Dataset statistics

```python
fig, axes = plt.subplots(1, 3, figsize=(18, 5))

df["quality"].value_counts().plot.bar(ax=axes[0], color="#2a9d8f")
axes[0].set_title("Quality tier"); axes[0].set_ylabel("images")

df["source_tag"].value_counts().head(10).plot.bar(ax=axes[1], color="#e76f51")
axes[1].set_title("Top source tags")

df["split"].value_counts().reindex(["train", "validation", "test"]).plot.bar(ax=axes[2], color="#264653")
axes[2].set_title("Recommended split")

for ax in axes:
    ax.tick_params(axis="x", rotation=45)
plt.tight_layout(); plt.show()
```

## 6. Use the recommended train / validation / test split

The split is stored per-image in the `split` column (deterministic 80 / 10 / 10).

```python
train = grounds.filter(lambda r: r["split"] == "train")
val   = grounds.filter(lambda r: r["split"] == "validation")
test  = grounds.filter(lambda r: r["split"] == "test")
print(f"train={len(train)}  val={len(val)}  test={len(test)}")
```
